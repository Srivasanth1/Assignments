```python
#importing pandas to read the csv
import pandas as pd
```

```python
dataset = pd.read_csv("experience_salary_data.csv")
```

```python
dataset
```

```python
#splitting input from the dataset
independent = dataset[["Experience (Years)"]]
independent
```

```python
#Splitting output from the data set
dependent = dataset[["Salary (USD)"]]
dependent
```

```python
#splitting the dataset in the ratio of 0.7 : 0.3
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Training the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight = regressor.coef_
```

```python
weight
```

```python
bias = regressor.intercept_
bias
```

```python
#getting value from predict xtest data.
Y_predict = regressor.predict(X_test)
```

```python
#Calculating r2 score
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_predict)
r_score
```

```python
import pickle
```

```python
filename = "Salary_predictor.sav"
```

```python
#Save the created model
pickle.dump(regressor, open(filename, "wb"))
```

```python
#load the saved file to get results
loaded_model = pickle.load(open(filename, "rb"))
```
